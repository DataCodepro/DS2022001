```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
sns.displot(x=tip['total_bill'],kde=True)
```

```python
sns.regplot(x = tip['total_bill'],y = tip['tip'])
```

```python
sns.regplot(x = tip['total_bill'],y = tip['size'])
```

```python
sns.jointplot(x = tip['total_bill'],y = tip['tip'])
```

```python
sns.pairplot(tip)
```

```python
sns.countplot(x =tip['day'])
```

```python
sns.barplot(x=tip['sex'],y=tip['total_bill'])
```

```python
sns.barplot(x=tip['sex'],y=tip['total_bill'],hue =tip['time'],palette='coolwarm')
```

```python
sns.swarmplot(x=tip['sex'],y=tip['total_bill'],hue =tip['time'],palette='coolwarm')
```

```python
sns.violinplot(x=tip['sex'],y=tip['total_bill'],hue =tip['time'],palette='coolwarm',split=True)
```

```python
sns.stripplot(x=tip['sex'],y=tip['total_bill'],hue =tip['time'],palette='Accent')
```

```python
sns.boxplot(y = tip['total_bill'])
```

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
df = pd.pivot_table(flight,index = 'year',values='passengers',columns = 'month')
df
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df)
```

```python
plt.figure(figsize=(10,10))
sns.clustermap(df)
```

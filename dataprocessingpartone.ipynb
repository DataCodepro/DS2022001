```python
month = ['January','Februray','March','April','May']
days = np.arange(1,31)
np.random.seed(5)
readings = np.random.randint(18,40,150)
readings = readings.reshape(30,5)

df_temp = pd.DataFrame(data = readings,index=days, columns = month)
```

```python
df_temp['Monthsum'] = df_temp['January'] + df_temp['Februray'] +df_temp['March'] +df_temp['April'] + df_temp['May']
# creating a column from existing columns in the dataframe
df_temp
```

```python
df_temp.loc['Average'] = (sum(df_temp['January']/len(df_temp)),sum(df_temp['Februray']/len(df_temp)),sum(df_temp['March']/len(df_temp)),sum(df_temp['April']/len(df_temp)),sum(df_temp['May']/len(df_temp)),sum(df_temp['Monthsum']/len(df_temp)))
```

```python
df_temp
```

```python
df_temp.drop('Monthsum',axis=1,inplace= True) #droping a column
```

```python
df_temp
```

```python
df_temp.drop('Average',axis=0,inplace= True) #droping a row
```

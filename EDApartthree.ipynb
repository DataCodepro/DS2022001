```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
studentper = pd.read_csv('StudentsPerformance.csv')
studentper.head()
```

```python
fig = px.histogram(x=studentper['math score'],text_auto=True,title='Distrubution of math score across all race',
                        marginal='violin',labels={'x':'math score','y':'Count'})
fig.show()
```

```python
df = studentper.iloc[:50,:]
fig =px.bar(x=df['race/ethnicity'],y =df['reading score'],color=df['parental level of education'],
                            title = 'Reading score across race and ethenicity and their qualifications',
                            text =df['reading score'],labels={'x':'Race','y':'Reading score'})
fig.show()
```

```python
fig = px.pie(names=studentper['race/ethnicity'],color=studentper['race/ethnicity'],
                            title='Percentage rate of each race')
fig.show()
```

```python
fig = px.sunburst(studentper, path=['race/ethnicity', 'parental level of education'], values='math score', 
                  title='Various race level of education  against math score')
fig.show()
```

```python
fig = px.box(studentper,y ='math score')
fig.show()
```
